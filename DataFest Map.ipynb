```python
import folium
import pandas as pd
import json
from folium import plugins

df = pd.read_csv('data2020_v4.csv')
df = df[['CITY', 'TOTAL_CRIME', 'DM', 'CHILD', 'zip_code']]
df = df.fillna(0)
df[['zip_code']] = df[['zip_code']].astype(int)

df[['zip_code']] = df[['zip_code']].astype(str)
df.loc[531, 'DM'] = 111
df
```

```python
df1 = pd.DataFrame()
df1['zipcode'] = [i for i in df['zip_code']]
df1['TOTAL_CRIME'] = [i for i in df['TOTAL_CRIME']]
df1

#initialize the LA County map
laMap = folium.Map(location=[34.0522,-118.2437], tiles = 'Stamen Toner', zoom_start = 9)
laMap.choropleth(data = df1, geo_data = 'laZips.geojson', columns=['zipcode', 'TOTAL_CRIME'],
                 key_on='feature.properties.zipcode',  fill_color='YlGn', fill_opacity = 1)
laMap.save('laChoropleth.html')
```

```python
df2 = pd.DataFrame()
df2['zipcode'] = [i for i in df['zip_code']]
df2['DM'] = [i for i in df['DM']]
df2

#initialize the LA County map
laMap = folium.Map(location=[34.0522,-118.2437], tiles = 'Stamen Toner', zoom_start = 9)
laMap.choropleth(data = df2, geo_data = 'laZips.geojson', columns=['zipcode', 'DM'],
                 key_on='feature.properties.zipcode', fill_color='YlGn', fill_opacity = 1)
laMap.save('laChoropleth_DM.html')
```
